```python
# q16 : Write a fun which will take input as a dict and give me out as a list of all the values 
# even in case of 2 level nesting it should work . 

d = {'k1' :"value" , "k2" : "values " ,"k3" : { "k12" : "sudh" , "k13"  : "gafasd"}}
def q16(di):
    l = []
    for v in di.values():
        if type(v) != dict:
            l.append(v)
        elif type(v) == dict:
            for v2 in v.values():
                l.append(v2)
    return l
q16(d)


```
