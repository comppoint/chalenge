```python
import mysql.connector as conn
import logging
import sys
import pandas as pd
```

```python
def create_logger(name, log_file, level=logging.INFO):
    handler = logging.FileHandler(log_file)
    handler.setFormatter(logging.Formatter('%(asctime)s %(levelname)s %(name)s %(message)s'))
    logger = logging.getLogger(name)
    logger.setLevel(level)
    logger.addHandler(handler)
    return logger
```

```python
logger = create_logger("Root", "./log/code.log")
db_logger = create_logger("MySQL", "./log/db.log")
```

```python
def run_query(query, database=None, host="localhost", user="root", password="Meena@123"):
    try:
        if database is None:
            logger.info("Connecting to MySQL")
            db_logger.info("Connecting to MySQL")
            db = conn.connect(host=host, user=user, password=password)
        else:
            logger.info(str("Connecting to Database " + database))
            db_logger.info(str("Connecting to Database " + database))
            db = conn.connect(host=host, user=user, password=password, database=database)
        cursor = db.cursor()
        logger.info("Executing " + query)
        db_logger.info("Executing " + query)
        cursor.execute(query)
        logger.info("Query run successfully " + query)
        db_logger.info("Query run successfully " + query)
        print(cursor.rowcount)
        if cursor.rowcount > 0:
            db.commit()
        else:
            result = cursor.fetchall()
            return result
    except conn.Error:
        a, b, c = sys.exc_info()
        db_logger.error(b)
    finally:
        db.close()
```

#### 2. Do a bulk load for these two table for respective dataset

#### 9. Try to find out total dress sell for individual dress id

```python
data = run_query("show columns from dress_data",database="dress")
query_data = ""
for i in data:
    if i[0] != "Dress_ID":
        query_data = query_data + "`" + i[0] + "`+"
query_data = query_data.rstrip("+")
query_data
```

```python
data = run_query("select Dress_ID, "+ query_data +"as total from dress_data",database="dress")
for i in data:
    print("Dress Id: {id} Total sell: {sell}".format(id=i[0],sell=i[1]))
```

#### 10. Try to find out a third highest most selling dress id

```python
data = run_query("select Dress_ID, "+ query_data +"as total from dress_data order by total desc LIMIT 1 OFFSET 2",database="dress")
for i in data:
    print("Dress Id: {id} Total sell: {sell}".format(id=i[0],sell=i[1]))
```
