```python
import pandas as pd
```

```python
perf = pd.read_csv("AgentPerformance.csv")
perf.head()
```

```python
login = pd.read_csv("AgentLogingReport.csv")
login.head()
```

```python
perf["Date"] = pd.to_datetime(perf["Date"])
perf.dtypes
```

#### 1 .Find out there avarage rating on weekly

```python
perf["week"] = perf["Date"].dt.isocalendar().week
perf.head()
```

```python
perf.groupby(["Agent Name","week"])["Average Rating"].mean()
```

#### 2 .Total working days for each agents

```python
perf[perf["Total Chats"]>0].groupby(["Agent Name"])["Agent Name"].count().sort_values()
```

#### 3. Total query that you hvae taken 

```python
perf.groupby(["Agent Name"])[["Total Chats"]].sum()
```

#### 4. total Feedback that you have received 

```python
print("total feedback received:",str(perf["Total Feedback"].sum()))
```

#### 5. a agent name who have average rating between 3.5 to 4

```python
data = perf.groupby(["Agent Name"])[["Average Rating"]].mean()
data[(data["Average Rating"] >= 3.5) & (data["Average Rating"] <= 4.0)]
```

#### 6 . Agent name who have rating lesss then 3.5

```python
data[data["Average Rating"] <3]
```

#### 7 . agent name who have rating more then 4.5

```python
data[data["Average Rating"] >= 4.5]
```

#### 9 . average weekly response time for each agent 

```python
perf["Average Response Time"] = pd.to_datetime(perf["Average Response Time"],format="%H:%M:%S",infer_datetime_format=True)
```

```python
pd.DataFrame(perf.groupby(["Agent Name","week"])["Average Response Time"].mean().dt.time)
```

#### 10 . average weekely resolution time for each agents 

```python
perf["Average Resolution Time"] = pd.to_datetime(perf["Average Resolution Time"],format="%H:%M:%S")
pd.DataFrame(perf.groupby(["Agent Name","week"])["Average Resolution Time"].mean().dt.time)
```

#### 11 . list of all agents name 

```python
perf["Agent Name"].unique()
```

#### 12 . percentage of chat on which they have received a feedback 

```python
data = perf.groupby("Agent Name")[["Total Chats","Total Feedback"]].sum()
```

```python
data["percnet"]=(data["Total Feedback"]/data["Total Chats"])*100
data
```

#### 13 . Total contributation hour for each and every agents weekly basis

```python
login["Date"] = pd.to_datetime(login["Date"])
login["week"] = login["Date"].dt.isocalendar().week
login
```

```python
login["Duration"] = pd.to_timedelta(login["Duration"])
pd.DataFrame(login.groupby(["week","Agent"])["Duration"].sum())
```

#### 14. total percentage of active hour for a month 

```python
login["month"] = login["Date"].dt.month
login
```

```python
pd.DataFrame(login.groupby(["month","Agent"])["Duration"].sum())
```
